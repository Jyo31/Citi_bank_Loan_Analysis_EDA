# Data Cleaning

```python
unnecessary_columns=['id','member_id']
loan.drop(unnecessary_columns,axis=1,inplace=True)
```

```python
loan.isnull().sum()
```

```python
loan1=loan.dropna(axis=0, how='any').copy()
```

```python
loan1.isnull().sum()
```

```python
loan1
```

```python
loan1['int_rate']=loan1['int_rate'].astype('string')
```

```python
loan1['int_rate']=loan1['int_rate'].apply(lambda x:x.split('%')[0])
```

```python
loan1['int_rate']=loan1['int_rate'].astype('float')
```

```python
loan1['int_rate']
```

```python
#here we are creating a function which will create a range in the interest rate provided to us.
#0-8: Low int rate
#8.1-13: Medium int rate
#>13: high int rate

def int_rate_binning(x):
    if x<8:
        return "low"
    elif x>8 and x<13:
        return "medium"
    else:
        return "high"
        
loan1['int_rate_category']=loan1['int_rate'].apply(lambda x:int_rate_binning(x))
```

```python
loan1['int_rate_category']
```

```python
loan1['annual_inc']
```

```python
#Here we are creating the function which will differiante between different income segment.
#low-income: 0-50000
#medium-income : 50001-100000
#high-income:100001>

def annual_inc_binning(x):
    if x<50000:
        return "low income"
    elif x>50000 and x< 100000:
        return "medium income"
    else:
        return "high income"
    
loan1['annual_inc_category']=loan1['annual_inc'].apply(lambda x:annual_inc_binning(x))
```

```python
loan1["annual_inc_category"]
```

```python
#loan_1['issue_d']--> month ,year -2011

loan1['month']=loan1['issue_d'].apply(lambda x:x.split('-')[0])
loan1['year']=loan1['issue_d'].apply(lambda x:'20'+ x.split('-')[1])
```

```python
loan1['year']
```

```python
loan1['month']
```

```python
loan1
```

```python
#Here we are trying to remove 'years' from the emp_length column and by using regular expression, we are trying to clean the values.

loan1['emp_level']=loan1['emp_length'].apply(lambda x:str(x).split(" y")[0])
```

```python
loan1['emp_level']
```

```python
import re
loan1['emp_level1']=loan1['emp_level'].apply(lambda x:re.split(r'<|>',x)[0])
```

```python
loan1['emp_level1']
```

```python
#Categorise based on the exp groups
def a(x):
    if x=="10+ years":
        return "11"
    
    if x<'5':
        return "Beginner"
    elif x>'5' and x<'11':
        return "Intermediate"
    else:
        return "Expert"
```

```python
loan1['emp_level1']=loan1['emp_level1'].apply(lambda x: a(x))
```

```python
loan1
```

```python
loan1['int_rate']
```

```python
def int_rate_binning(x):
    if x<=8.0:
        return 'low_interest'
    elif x>8.0 and x <= 13.0:
        return 'medium_interest'
    else:
        return 'high_interest'
```

```python
loan1['int_rate_category'] = loan1['int_rate'].apply(lambda x : int_rate_binning(x))
```
